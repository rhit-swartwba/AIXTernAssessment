```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('XTern 2024 Artificial Intelegence Data Set.xlsx')

data.head()
```

```python
num_cols = 2
num_rows = (len(data.columns) + 1) // num_cols  # +1 to handle an odd number of columns

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 18))
fig.subplots_adjust(wspace=0.5, hspace=0.5)

for i, column in enumerate(data.columns):
    row_idx = i // num_cols
    col_idx = i % num_cols

    if data[column].dtype in [int, float]:
        data[column].plot(kind='hist', bins=20, edgecolor='black', ax=axes[row_idx, col_idx])
        axes[row_idx, col_idx].set_title(f'Distribution of {column}')
        axes[row_idx, col_idx].set_xlabel(column)
    else:
        data[column].value_counts().plot(kind='bar', ax=axes[row_idx, col_idx])
        axes[row_idx, col_idx].set_title(f'Distribution of {column}')
        axes[row_idx, col_idx].set_xlabel(column)
        axes[row_idx, col_idx].set_ylabel('Count')
        axes[row_idx, col_idx].tick_params(axis='x', rotation=45)

for i in range(len(data.columns), num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```
